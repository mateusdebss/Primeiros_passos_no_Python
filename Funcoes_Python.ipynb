```python
# Bliotecas
import pandas as pd
import seaborn as sns
```

```python
# Dados
tabela = sns.load_dataset("iris").tail(10)
tabela
```

```python
# Código manual e repetitivo
tabela.sepal_length = (tabela.sepal_length - tabela.sepal_length.min(skipna=True)) / (
    tabela.sepal_length.max(skipna=True) - tabela.sepal_length.min(skipna=True))
tabela.sepal_length
```

```python
# Replicando manualmente paras as demais colunas
tabela.sepal_width = (tabela.sepal_width - tabela.sepal_width.min(skipna=True)) / \
    (tabela.sepal_width.max(skipna=True) - tabela.sepal_width.min(skipna=True))
tabela.petal_length = (tabela.petal_length - tabela.petal_length.min(skipna=True)) / (
    tabela.petal_length.max(skipna=True) - tabela.petal_length.min(skipna=True))
tabela.petal_width = (tabela.petal_width - tabela.petal_width.min(skipna=True)) / \
    (tabela.petal_width.max(skipna=True) - tabela.petal_width.min(skipna=True))
tabela
```

```python
# Código limpo e organizado
## def nome_funcao(arg1, arg2, ...):
    # corpo: código que a função executa
    ##return  # o objeto retornado

# Definir função de normalização dos dados
def normalizar(x):
    x_min = x.min(skipna=True)
    return (x - x_min) / (x.max(skipna=True) - x_min)
```

```python
# Utilizando a função definida
tabela = sns.load_dataset("iris").tail(10)
tabela.sepal_length = normalizar(tabela.sepal_length)
tabela.sepal_width = normalizar(tabela.sepal_width)
tabela.petal_length = normalizar(tabela.petal_length)
tabela.petal_width = normalizar(tabela.petal_width)
tabela
```
